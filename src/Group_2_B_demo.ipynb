## Constants

```python
TEST_PATH = '.data/snli/snli_1.0/test.csv'
RESULTS_PATH = '.data/snli/snli_1.0/test_with_labels.jsonl'
MODEL_PATH = 'results/model.pt'

LABEL = 'label'

SEED = 1234
EVALUATION_BATCH_SIZE = 16
```

## Preprocess Datasets

```python
with open(TEST_PATH, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        with jsonlines.open(RESULTS_PATH, mode='w') as writer:
            for index, row in enumerate(csv_reader):
                row[LABEL] = "0"
                writer.write(row)
```
